```python
df = pd.read_excel(
	'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx',
	sheet_name='Canada by Citizenship',
	skiprows=range(20),
	skipfooter=2,
	engine='openpyxl'
)
```

```python
df.head()
```

```python
df = df.set_index('OdName')
```

```python
df
```

```python
df_can = df.loc[:, np.arange(1980, 2014).tolist()]
df_can
```

```python
fig, ax = plt.subplots(2,2, figsize=(20, 15))
ax[0, 0].plot(df_can.columns, df_can.loc['Afghanistan'].values)
ax[0, 0].set_ylabel('number of immigrants')
ax[0, 0].set_xlabel('years')
ax[0, 0].set_title('Afghanistan move to Canada')

ax[0, 1].plot(df_can.columns, df_can.loc['Viet Nam'].values)
ax[0, 1].set_ylabel('number of immigrants')
ax[0, 1].set_xlabel('years')
ax[0, 1].set_title('Viet Nam move to Canada')

ax[1, 0].plot(df_can.columns, df_can.loc['Albania'].values)
ax[1, 0].set_ylabel('number of immigrants')
ax[1, 0].set_xlabel('years')
ax[1, 0].set_title('Albania move to Canada')

ax[1, 1].plot(df_can.columns, df_can.loc['Algeria'].values)
ax[1, 1].set_ylabel('number of immigrants')
ax[1, 1].set_xlabel('years')
ax[1, 1].set_title('Algeria move to Canada')

plt.show()
```

```python
tmp_df = df_can.loc[['Afghanistan', 'Viet Nam', 'Albania', 'Algeria'], :]
tmp_df
```

```python
tmp_df.transpose().plot(kind='area', stacked=False, alpha=0.5)
plt.ylabel('number of immigrants')
plt.xlabel('years')
plt.title('Move to Canada from 1980 to 2013')
plt.show()
```

```python
ax = tmp_df.transpose().plot(kind='area', stacked=False, alpha=0.5)
ax.set_ylabel('number of immigrants')
ax.set_xlabel('years')
ax.set_title('Move to Canada from 1980 to 2013')
plt.show()
```

```python
df_continent = df.loc[:, ['RegName'] + np.arange(1980, 2014).tolist()]
df_continent = df_continent.groupby(by=['RegName'], axis=0).sum()
df_continent
```

```python
df_continent.iloc[[1,2,4], :].transpose().plot(kind='box')
plt.xlabel('Areas Name')
plt.ylabel('Number of immigrants per year')
plt.title('Annual immigrants from 1980 to 2013')
plt.show()
```
